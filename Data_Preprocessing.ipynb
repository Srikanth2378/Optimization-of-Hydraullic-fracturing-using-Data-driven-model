Dropping features that are highly correlated to get stable model. Regression model tries to establish relationship with independent variables to get a good model to predict. To get stable model we need to drop highly correlated features

```python
# lists to store values
ML_Model = []
R2 = []
MAE=[]
RMSE=[]

#function to call for storing the results
def storeResults(model,a,b,c):
    ML_Model.append(model)
    R2.append(a)
    MAE.append(b)
    RMSE.append(c)
```

**Linear Regession**

```python
Y=df['Production']
X= df.drop(['Production','Boggess Pad'],axis=1)

print("Shape Of X:",X.shape)
print("Shape of Y:",Y.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics

#Split data to training and testing #RANDOMLY 
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.25, random_state=224)  #80/20 or 70/30 depends on how much data you want to distribute and sample size. After building the model you can go back and try different split.
```

```python
#Create a linear regression model lr
lr=LinearRegression()
#fit and train the lr model
lr_fit = lr.fit(X_train,Y_train)
#predict the Aroot(k) from training data
Y_train_pred_lr=lr_fit.predict(X_train)
#predict the Aroot(k) from testing data
Y_test_pred_lr=lr_fit.predict(X_test)
```

```python
print("R^2 on training set: %f" % lr_fit.score(X_train, Y_train))
#R2
R2_lr_ts= lr_fit.score(X_test, Y_test)
print("R^2 on test set:",R2_lr_ts)
```

```python
#MAE- Mean Absolute Error
MAE_lr_tr = metrics.mean_absolute_error(Y_train,Y_train_pred_lr)
print("MAE for trainig:",MAE_lr_tr)
MAE_lr_ts = metrics.mean_absolute_error(Y_test,Y_test_pred_lr)
print("MAE for testing:",MAE_lr_ts)
print("\n")
#MSE- Mean Squared Error
MSE_lr_tr = metrics.mean_squared_error(Y_train,Y_train_pred_lr)
print("MSE for trainig:",MAE_lr_tr)
MSE_lr_ts = metrics.mean_squared_error(Y_test,Y_test_pred_lr)
print("MSE for testing:",MAE_lr_ts)
print("\n")

#RMSE - Root Mean Squared Error
RMSE_lr_tr=np.sqrt(metrics.mean_squared_error(Y_train,Y_train_pred_lr))
RMSE_lr_ts=np.sqrt(metrics.mean_squared_error(Y_test,Y_test_pred_lr))
print("RMSE for Training Data predictions: ",RMSE_lr_tr)
print("RMSE for Testing Data predictions: ",RMSE_lr_ts)
```

```python
#Storing the values
storeResults('Linear Regression',R2_lr_ts,MAE_lr_ts,RMSE_lr_ts)
```
